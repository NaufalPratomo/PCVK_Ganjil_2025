# **Praktikum**

Langkah 1

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Langkah 2

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"]
#baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert color to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #show image
  print('\n')
```

Langkah 3

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

Langkah 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
    image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

4. Perhatikan potongan kode progam berikut :
Apakah kegunaan kode [255,255,255] ? Jelaskan!
*   Kode [255, 255, 255] merepresentasikan nilai warna putih dalam model warna BGR (Format default OpenCV).

*   Dalam model warna BGR 8-bit, setiap channel (Blue, Green, Red) memiliki nilai dari 0 hingga 255.

*   [255, 255, 255] berarti nilai maksimum untuk semua channel (Blue, Green, dan Red), yang ketika dicampur akan menghasilkan warna putih.

Jadi, kode tersebut digunakan untuk mengubah warna pixel pada posisi tertentu menjadi putih.

5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!

*   Pixel adalah elemen terkecil dalam sebuah citra digital, yang menyimpan informasi warna.

*   Resolusi mengacu pada jumlah total pixel yang menyusun sebuah citra, biasanya dinyatakan dalam dimensi lebar x tinggi (misalnya, 1920x1080 pixel).

Keterkaitan:

*   Resolusi Tinggi: Berarti jumlah pixel lebih banyak dalam area yang sama. Ini menghasilkan citra yang lebih detail, tajam, dan halus karena lebih banyak informasi yang disimpan. Namun, ukuran filenya juga lebih besar.

*   Resolusi Rendah: Berarti jumlah pixel lebih sedikit. Ini menyebabkan citra terlihat kasar, blur, atau kotak-kotak (pixelated) ketika diperbesar, karena kurang detail. Ukuran file lebih kecil.

Intinya, resolusi adalah ukuran kuantitas pixel, sedangkan kualitas tampilan citra sangat bergantung pada resolusi ini.



# **TUGAS**

1. Lakukan langkah-langkah praktikum seperti diatas

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"]
#baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert color to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #show image
  print('\n')
```

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
img_asli1 = image.copy()
img_asli2 = image.copy()
img_vertikal = image.copy()
img_diagonal = image.copy()

# garis vertikal
for x in range(tinggi):
    img_vertikal[x, lebar//2] = [255,255,255]

# garis silang (X)
# Garis diagonal kiri atas ke kanan bawah
for x in range(lebar):
    y_diag1 = int((tinggi/lebar) * x)
    if 0 <= y_diag1 < tinggi:
        img_diagonal[y_diag1, x] = [255, 255, 255]

# Garis diagonal kanan atas ke kiri bawah
for x in range(lebar):
    y_diag2 = int(tinggi - 1 - (tinggi/lebar) * x)
    if 0 <= y_diag2 < tinggi:
        img_diagonal[y_diag2, x] = [255, 255, 255]

# susun jadi grid 2x2
row1 = cv.hconcat([img_asli1, img_vertikal])
row2 = cv.hconcat([img_asli2, img_diagonal])
final_grid = cv.vconcat([row1, row2])

# tampilkan
cv2_imshow(final_grid)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu

```python
# Salin gambar asli untuk diproses
image_tugas3 = image_2.copy()

# Tentukan panjang garis (50% dari lebar gambar)
panjang_garis = int(lebar * 0.5)
start_x = int((lebar - panjang_garis) / 2)
end_x = start_x + panjang_garis

# Buat garis horizontal di tengah dengan panjang tertentu
for x in range(start_x, end_x):
    image_tugas3[int(tinggi/2), x] = [255, 255, 255]  # Warna putih

# Gabungkan gambar asli dan gambar yang sudah diproses
final_frame_tugas3 = cv.hconcat([image_2, image_tugas3])

# Tampilkan hasil
cv2_imshow(final_frame_tugas3)
print(f"Garis horizontal dari x={start_x} sampai x={end_x} (panjang: {panjang_garis} pixel)")
print(f"Posisi y: {int(tinggi/2)} (tengah gambar)")
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar

```python
img_asli = image.copy()
img_kotak = image.copy()

# koordinat kotak (x1,y1) ke (x2,y2)
x1, y1 = 50, 50
x2, y2 = 150, 150

for x in range(x1, x2):
    for y in range(y1, y2):
        img_kotak[x, y] = [255,255,255]

final_frame = cv.hconcat([img_asli, img_kotak])

cv2_imshow(final_frame)
```
