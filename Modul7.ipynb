# Praktikum Filter

**Persiapan Library dan Memuat Gambar**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math

# Langkah d: Muat citra dan ubah menjadi citra keabuan (grayscale)
try:
    # Membaca gambar dari file
    img = cv2.imread('gambar1.tiff')
    # Konversi dari BGR (standar OpenCV) ke RGB untuk ditampilkan dengan benar oleh Matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Konversi ke grayscale untuk proses konvolusi
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
except Exception as e:
    print(f"Error: Gagal memuat gambar 'gambar1.tiff'. Pastikan file sudah diunggah.")
    # Membuat gambar pengganti jika file tidak ditemukan, agar kode tetap bisa berjalan
    img_rgb = np.zeros((256, 256, 3), dtype=np.uint8)
    img_gray = np.zeros((256, 256), dtype=np.uint8)

# Menampilkan citra asli dan citra grayscale
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_rgb)
axes[0].set_title('Citra Asli (RGB)')
axes[0].axis('off')

axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Citra Grayscale')
axes[1].axis('off')

plt.show()
```

**Membuat Fungsi Konvolusi**

```python
# Langkah c: Membuat fungsi konvolusi dari awal
def convolution2d(image, kernel):
    """
    Melakukan operasi konvolusi 2D pada citra grayscale.

    Args:
      image (np.ndarray): Citra input dalam format grayscale.
      kernel (np.ndarray): Matriks kernel untuk filter.

    Returns:
      np.ndarray: Citra hasil konvolusi.
    """
    # Mendapatkan dimensi kernel dan citra
    k_height, k_width = kernel.shape
    img_height, img_width = image.shape

    # Menghitung ukuran padding yang dibutuhkan
    # Padding ini memastikan titik tengah kernel bisa menjangkau setiap piksel citra
    pad_h = k_height // 2
    pad_w = k_width // 2

    # Membuat citra baru dengan padding nol di sekelilingnya
    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant', constant_values=0)

    # Menyiapkan matriks output dengan ukuran yang sama seperti citra input
    output_image = np.zeros_like(image, dtype='float32')

    # Melakukan proses konvolusi dengan menggeser kernel ke seluruh citra
    for y in range(img_height):
        for x in range(img_width):
            # Mengambil bagian citra (region of interest) yang tumpang tindih dengan kernel
            roi = padded_image[y : y + k_height, x : x + k_width]

            # Menghitung hasil konvolusi: perkalian elemen-demi-elemen lalu dijumlahkan
            pixel_value = np.sum(roi * kernel)
            output_image[y, x] = pixel_value

    return output_image
```

**Mendefinisikan Kernel dan Menerapkan Filter**

**1. Filter Sharpen**

```python
# Kernel untuk Sharpen
kernel_sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

# Terapkan konvolusi
sharpened_image = convolution2d(img_gray, kernel_sharpen)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')

# Hasilnya perlu di-clip agar nilainya tetap di rentang 0-255 untuk ditampilkan
axes[1].imshow(np.clip(sharpened_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Filter Sharpen')
axes[1].axis('off')

plt.show()
```

**2. Filter Average (Blur)**

```python
# Kernel untuk Average Filter
kernel_average = (1/9) * np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

# Terapkan konvolusi
average_image = convolution2d(img_gray, kernel_average)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(average_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Filter Average (Blur)')
axes[1].axis('off')

plt.show()
```

**3. Low Pass Filter**

```python
# Kernel untuk Low Pass Filter
kernel_lowpass = (1/12) * np.array([
    [1, 1, 1],
    [1, 4, 1],
    [1, 1, 1]
])
# Terapkan konvolusi
lowpass_image = convolution2d(img_gray, kernel_lowpass)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(lowpass_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Low Pass Filter')
axes[1].axis('off')

plt.show()
```

**4. High Pass Filter**

```python
# Kernel untuk High Pass Filter
kernel_highpass = np.array([
    [-1, 0, 1],
    [-1, 0, 3],
    [-3, 0, 1]
])
# Terapkan konvolusi
highpass_image = convolution2d(img_gray, kernel_highpass)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(highpass_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil High Pass Filter')
axes[1].axis('off')

plt.show()
```

**5. Filter Emboss**

```python
# Kernel untuk Emboss
kernel_emboss = np.array([
    [-2, -1, 0],
    [-1,  1, 1],
    [ 0,  1, 2]
])

# Terapkan konvolusi
emboss_image = convolution2d(img_gray, kernel_emboss)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(emboss_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Filter Emboss')
axes[1].axis('off')

plt.show()
```

**6. Left Sobel Edge Detection**

```python
# Kernel untuk Left Sobel Edge Detection
kernel_left_sobel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

# Terapkan konvolusi
left_sobel_image = convolution2d(img_gray, kernel_left_sobel)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(left_sobel_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Left Sobel Edge Detection')
axes[1].axis('off')

plt.show()
```

**7. "Canny" Edge Detection (Laplacian Kernel)**

```python
# Kernel Laplacian (disebut Canny di modul)
kernel_canny_laplacian = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# Terapkan konvolusi
canny_laplacian_image = convolution2d(img_gray, kernel_canny_laplacian)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(canny_laplacian_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Deteksi Tepi (Laplacian)')
axes[1].axis('off')

plt.show()
```

**8. Prewitt Edge Detection**

```python
# Kernel untuk Prewitt Edge Detection (Horizontal)
kernel_prewitt = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# Terapkan konvolusi
prewitt_image = convolution2d(img_gray, kernel_prewitt)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(prewitt_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil Prewitt Edge Detection')
axes[1].axis('off')

plt.show()
```

**9. 5x5 Gaussian Blur**

```python
# Kernel untuk 5x5 Gaussian Blur
kernel_gaussian_5x5 = (1/256) * np.array([
    [1,  4,  6,  4, 1],
    [4, 16, 24, 16, 4],
    [6, 24, 36, 24, 6],
    [4, 16, 24, 16, 4],
    [1,  4,  6,  4, 1]
])

# Terapkan konvolusi
gaussian_5x5_image = convolution2d(img_gray, kernel_gaussian_5x5)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(gaussian_5x5_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil 5x5 Gaussian Blur')
axes[1].axis('off')

plt.show()
```

**10. 21x21 Gaussian Blur**

```python
# Membuat kernel Gaussian 21x21 secara otomatis
kernel_size = 21
sigma = math.sqrt(kernel_size)
gaussian_kernel_1d = cv2.getGaussianKernel(kernel_size, sigma)
kernel_gaussian_21x21 = gaussian_kernel_1d @ gaussian_kernel_1d.transpose()

# Terapkan konvolusi
gaussian_21x21_image = convolution2d(img_gray, kernel_gaussian_21x21)

# Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Grayscale Asli')
axes[0].axis('off')
axes[1].imshow(np.clip(gaussian_21x21_image, 0, 255), cmap='gray')
axes[1].set_title('Hasil 21x21 Gaussian Blur')
axes[1].axis('off')

plt.show()
```
